```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from statistics import mode
import math
import seaborn as sns
from matplotlib import pyplot as plt
%matplotlib inline
import random

#Import the Iris dataset using seaborn library's load_dataset function
iris=sns.load_dataset('iris')

#Print the first few rows of the iris dataset
print(iris.head())

#Find euclidean distances of each training data from test data and then return the neighbours for a particular record in testset
def findNeighbours(trainset,testRow,kValue):
    distance=0.0
    dist=list()
    neighbours=list()
    for index,trainRow in trainset.iterrows():
        for i in range(len(trainRow)-1):
            distance+=(testRow[i]-trainRow[i])**2
        dist.append({'index':index,'distance':math.sqrt(distance)})
        distance=0.0    
    #This would sort the dictionary inside the list in ascending order with respect to distance
    dist=sorted(dist,key=lambda k:k['distance'])
    for i in range(kValue):
        neighbours.append(dist[i])   
    return neighbours    

#Get the neighbours for each observation in the testset
def evaluate_kNeighboursAlgo(trainset,testRows,kValue):
    neighbourData=list()
    for index,testRow in testRows.iterrows():
        neighbourData.append(findNeighbours(trainset,testRow,kValue))
    return neighbourData
        
def stringToInt(dataset,column):
    class_values = [row[column] for index,row in dataset.iterrows()]
    unique = set(class_values)
    lookup = dict()
    for i, value in enumerate(unique):        
        lookup[value] = i
        print('[%s] => %d' % (value, i))
    for index,row in dataset.iterrows(): 
        row[column] = lookup[row[column]]    
    return lookup

#Seleting the target data column
target=iris['species']

#Split the data into training and test set
iris_dataTrain,iris_dataTest,iris_targetTrain,iris_targetTest=train_test_split(iris,target,test_size=0.3,random_state=30)
#print(iris_dataTrain.to_string())

#Drop the classifier column from the test data set
iris_dataTest=iris_dataTest.drop('species',axis=1)

#Create a lookup for class values
stringToInt(iris_dataTrain,iris_dataTrain.shape[1]-1)

#Evaluate the K-Nearest Neighbours algorithm for the test set and get the k neighbours for each row in test data
neighbourData=evaluate_kNeighboursAlgo(iris_dataTrain,iris_dataTest,3)

#print(neighbourData)

output_value=list()

#print(iris_dataTrain.loc[neighbourData[0][0]['index'],'species'])

#Get the respective class labels for each neighbour  
for i in range(len(neighbourData)):
    for j in range(len(neighbourData[i])): 
        output_value.append({'index':i,'value':iris_dataTrain.loc[neighbourData[i][j]['index'],'species']}) 
#print(output_value)

op=pd.DataFrame(output_value)

#print(op.groupby('index').agg(lambda x:x.value_counts().index[0])['value'])

pred=list()

#Print the predicted output label for the given test data by finding the mode from the output values
#Mode is used to find the predicted label as this is a classification problem
#For Regression problems, mean would be calculated for prediction

#value_counts() function returns the mode of the data
pred.append(op.groupby('index').agg(lambda x:x.value_counts().index[0])['value'])

print("\nThe Predictions \n")
print(pred)

#print(accuracy_score(iris_targetTest,pred,normalize=True))

print(iris_targetTest.to_string())
```
